Next Section: Calculating total number of years in each of the possible states (A, B*, B, D)

```python
timetotalA = 0              # Initializing variables that will store the total number of years spent in a given state
timetotalStar = 0
timetotalB = 0
timetotalD = 0

for i in range(200):
  for j in range(3):
    if df.iloc[i,(2*j + 1)] == 'A':
     timetotalA = timetotalA + (df.iloc[i,(2*j + 2)] / 12)
    elif df.iloc[i,(2*j + 1)] == '*':
     timetotalStar = timetotalStar + (df.iloc[i,(2*j + 2)] / 12)
    elif df.iloc[i,(2*j + 1)] == 'B':
     timetotalB = timetotalB + (df.iloc[i,(2*j + 2)] / 12)
    elif df.iloc[i,(2*j + 1)] == 'D':
     timetotalD = timetotalD + (df.iloc[i,(2*j + 2)] / 12)
```

The next section creates a list of 24 characters for each of the firms. Each character represesents a month of rating. For example, A***BBBBBBBBBBBBBBBBBBBB means that the firm spent the first month of the observed 24-month period in state A, then three months in B* and the rest in B.

```python
list_1 = []                        # Creating an empty list which will serve as a list of lists

for i in range(0, 200):

  list_2 = df.iloc[i,1] * df.iloc[i,2] + df.iloc[i,3] * df.iloc[i,4] + df.iloc[i,5] * df.iloc[i,6]     # Creating a list with the ratings of the firm number i
  list_1.append(list_2)            # Appending the list of lists. Each element on the list is a list with the ratings of a firm.
```

```python
list_3 = []                                 # Creating an empty list

for i in range(0,200):                      # Fist for loop with 200 iterations for the 200 firms
 s1 = str(list_1[i][0]) + str(list_1[i][11])      # Creating a pair of rating classifications that are in 12 months distance
 s2 = str(list_1[i][12]) + str(list_1[i][23])
 list_3.append(s1)
 list_3.append(s2)
```

```python
AA = 0
Astar = 0
AB = 0
AD = 0
starA = 0
starstar = 0
starB = 0
starD = 0
BA = 0
Bstar = 0
BB = 0
BD = 0
DD = 0

for i in range(len(list_3)):
  if list_3[i] == 'AA':
    AA = AA + 1
  elif  list_3[i] == 'A*':
    Astar = Astar + 1
  elif  list_3[i] == 'AB':
    AB = AB + 1
  elif  list_3[i] == 'AD':
    AD = AD + 1
  elif  list_3[i] == '*A':
    starA = starA + 1
  elif  list_3[i] == '**':
    starstar = starstar + 1
  elif  list_3[i] == '*B':
    starB = starB + 1
  elif  list_3[i] == '*D':
    starD = starD + 1
  elif  list_3[i] == 'BA':
    BA = BA + 1
  elif  list_3[i] == 'B*':
    Bstar = Bstar + 1
  elif  list_3[i] == 'BB':
    BB = BB + 1
  elif  list_3[i] == 'BD':
    BD = BD + 1
  elif  list_3[i] == 'DD':
    DD = DD + 1
print(AA+Astar+AB+AD+starA+starstar+starB+starD+BA+Bstar+BB+BD+DD)  # This is a quick check whether all possibilities pairs are counted. The result should be 400.
```

```python
import numpy as np
from scipy.linalg import logm, expm

total_A = AA + Astar + AB + AD
total_star = starA + starstar + starB + starD
total_B = BA + Bstar + BB + BD
total_D = DD

df1 = pd.DataFrame(index=["Initial State: A", "Initial State: B*", "Initial State: B", "Initial State: D"], columns=["A", "B*", "B", "D"])

df1.iloc[0,0] = AA
df1.iloc[0,1] = Astar
df1.iloc[0,2] = AB
df1.iloc[0,3] = AD
df1.iloc[1,0] = starA
df1.iloc[1,1] = starstar
df1.iloc[1,2] = starB
df1.iloc[1,3] = starD
df1.iloc[2,0] = BA
df1.iloc[2,1] = Bstar
df1.iloc[2,2] = BB
df1.iloc[2,3] = BD
df1.iloc[3,0] = 0         # This has to be zero, as D (Default) is an absorbing state, meaning that if a rating gets downgraded to D, it stays there; there is no move from D to A.
df1.iloc[3,1] = 0
df1.iloc[3,2] = 0
df1.iloc[3,3] = DD



df2 = df1.copy()
df3 = df1.copy()

df2.iloc[0]=df2.iloc[0]/total_A
df2.iloc[1]=df2.iloc[1]/total_star
df2.iloc[2]=df2.iloc[2]/total_B
df2.iloc[3]=df2.iloc[3]/total_D

df3.iloc[0]=df3.iloc[0]/timetotalA
df3.iloc[1]=df3.iloc[1]/timetotalStar
df3.iloc[2]=df3.iloc[2]/timetotalB
df3.iloc[3]=df3.iloc[3]/total_D

df3.iloc[0,0] = - (df3.iloc[0,1] + df3.iloc[0,2] + df3.iloc[0,3])
df3.iloc[1,1] = - (df3.iloc[1,0] + df3.iloc[1,2] + df3.iloc[1,3])
df3.iloc[2,2] = - (df3.iloc[2,0] + df3.iloc[2,1] + df3.iloc[2,3])
df3.iloc[3,3] =   (df3.iloc[3,0] + df3.iloc[3,1] + df3.iloc[3,2])
```

```python
total_A_poorer = AA + Astar + AB + AD
total_B_poorer = BA + Bstar + BB + BD + starA + starstar + starB + starD
total_D_poorer = DD

timetotalA_poorer = timetotalA
timetotalB_poorer = timetotalStar + timetotalB
timetotalD_poorer = 0

AA_poorer = AA
AB_poorer = Astar + AB
AD_poorer = AD

BA_poorer = BA + starA
BB_poorer = starstar + BB + starB + Bstar
BD_poorer = starD + BD

DD_poorer = DD

df4 = pd.DataFrame(index=["Initial State: A", "Initial State: B", "Initial State: D"], columns=["A", "B", "D"])

df4.iloc[0,0] = AA_poorer
df4.iloc[0,1] = AB_poorer
df4.iloc[0,2] = AD_poorer
df4.iloc[1,0] = BA_poorer
df4.iloc[1,1] = BB_poorer
df4.iloc[1,2] = BD_poorer
df4.iloc[2,0] = 0         # This has to be zero, as D (Default) is an absorbing state, meaning that if a rating gets downgraded to D, it stays there; there is no move from D to A.
df4.iloc[2,1] = 0
df4.iloc[2,2] = DD

df5 = df4.copy()
df6 = df4.copy()

df5.iloc[0]=df5.iloc[0]/total_A_poorer
df5.iloc[1]=df5.iloc[1]/total_B_poorer
df5.iloc[2]=df5.iloc[2]/total_D_poorer

df6.iloc[0]=df6.iloc[0]/timetotalA_poorer
df6.iloc[1]=df6.iloc[1]/timetotalB_poorer
df6.iloc[2]=df6.iloc[2]/timetotalB_poorer


df6.iloc[0,0] = - (df6.iloc[0,1] + df6.iloc[0,2])
df6.iloc[1,1] = - (df6.iloc[1,0] + df6.iloc[1,2])
df6.iloc[2,2] =   (df6.iloc[2,0] + df6.iloc[2,1])
```

3. For both state spaces, explain the difference in the 1-year default probabilities between the discrete-time and the continuous-time estimator.

Solution: A great weakness of the cohort method is that it is too dependent on actual, observed past changes in the rating. If historical data does not go back long enough in the past to have rare default events, it can be that the estimated default probability is zero which appears to suggest that the initial state is riskless. This can never happen with the continuous-time method, as it always produces positive transition probabilities.

4. What happens to the 1-year default probability from the "official" states
A and B when there is an additional state? Do the results correspond to
the interpretation of B* as an "excited" state?

Solution:

When there is an additional state, the probability of defaulting from A within a year does not change with the cohort method. This is due to the fact that there was only one actual observation of a firm going from A to D in the course of one year. As for BD of the cohort method, the probability decreases with increasing state space.

Continuous-Time method: If we replace all of the B* states with B, the firms will spend more time in B, so the elements of the generator matrix will decrease which will in turn lead to the decrease of the transition probability.

If one looks at the transition intensities of the generator matrices, it seems to be clear that these are greater than with other states, so B* seems to be an "excited" state.
